```python
import mimetypes
import requests

import urllib
import urllib.request
import itertools
import os
import cv2

import socket
timeout = 3617
socket.setdefaulttimeout(timeout)

from tqdm import tqdm_notebook
from miniutils.progress_bar import iparallel_progbar
import numpy as np
import time

from multiprocessing.pool import ThreadPool
```

```python
def loadImage(z):
    path, url, counter = z
#     time.sleep(0.1 + np.random.random() / 9)
    p = os.path.join(path, str(counter))
    try:
        with requests.get(url, timeout=10, stream=True) as r:
            content_type = r.headers['content-type']
            if r.status_code == 200 and content_type.split('/')[0] == 'image':
                ext = mimetypes.guess_extension(content_type)
                p = p + ext
                with open(p, 'wb') as f:
                    for chunk in r.iter_content(chunk_size=1024):
                        if chunk:
                            f.write(chunk)
    except Exception as e:
        pass
```

```python
def storeImages(src, main_folder="./data"):
    for url, path in tqdm_notebook(src):
        time.sleep(1)
        path = os.path.join(main_folder, path)
        if not os.path.exists(path):
            os.makedirs(path)
        try:
            image_urls = str(urllib.request.urlopen(url).read())
            image_urls = image_urls.split(r'\n')
            z = zip(
                itertools.repeat(path),
                image_urls,
                range(len(image_urls))
            )
            with ThreadPool(16) as p:
                p.map(loadImage, z)
        except Exception as e:
            pass
```

```python
storeImages(src)
```

```python
from skimage.io import *
from glob import glob
from pathlib import Path
%matplotlib inline

def is_image(x):
    ext = x.split('.')[-1]
    accepted = {'jp', 'png', 'gif'}
    for a in accepted:
        if a in ext.lower():
            return True
    return False

def removeCorruptedImages():
    paths = list(filter(is_image, (map(str, Path('./').glob('**/*.*')))))
    for path in tqdm_notebook(paths):
        try:
            x = imread(path)
        except:
            os.remove(path)
```

```python
import hashlib

BLOCK_SIZE = 65536

def get_file_hash(path, b_size=BLOCK_SIZE):
    with open(path, 'rb') as f:
        hasher = hashlib.sha256()
        b = f.read(b_size)
        while len(b):
            hasher.update(b)
            b = f.read(b_size)
    return hasher.hexdigest()
```

```python
from collections import defaultdict
def get_hashes_dict(paths, b_size=BLOCK_SIZE):
    d = defaultdict(list)
    for p in tqdm_notebook(paths):
        h = get_file_hash(p, b_size=b_size)
        d[h].append(p)
    return d
```

```python
d = get_hashes_dict(paths)
```

```python
from tqdm import trange

def find_paths_of_duplicated(d):
    duplicated = []
    duplicates = list(
        filter(
            lambda x: len(x[1]) > 1, 
            sorted(
                d.items(), 
                key=lambda x: len(x[1]), 
                reverse=True)
        )
    )
    for i in trange(len(duplicates)):
        duplicates[i][1].pop()
        duplicated.extend(duplicates[i][1])
    return duplicated
```

```python
duplicated = find_paths_of_duplicated(d)
```
